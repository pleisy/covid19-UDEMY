```python
today = date.today()
print(today)

#read the Dataset sheet. The dataset will be called "data".
path = %pwd
print("PATH:", path)
local = input("Données en local [Y] ou depuis le site WEB")
if local == "Y" or local == "y":
    print("En Local ...")
    file = "./data/covid19_" + str(today) + ".csv"
else:
    print("Téléchargement du fichier ... patientez")
    file = "https://opendata.ecdc.europa.eu/covid19/casedistribution/csv"
try:
    data = pd.read_csv(file)
    data.head()
except:
    print("file {} not found".format(file))
    
if local != "Y":
    data.to_csv("./data/covid19_" + str(today) + ".csv")
```

```python
print(data.shape)
countries = data["countriesAndTerritories"].unique()
countries2 = data["geoId"].unique()
print("Nombre pays:", len(countries))
print(countries2)
print("Somme des cas confirmés:", data["cases"].sum())
print("Somme des morts:", data["deaths"].sum())
```

```python

#pays = ["FR", "IT"]
pays = ["FR", "IT", "ES","DE", "US","IN","BR","MX"]
noms  = ["data_pays_" + pays[n] for n in range(len(pays))]
noms2 = ["data_pays_" + pays[n] + "_2" for n in range(len(pays))]

for n in range(len(pays)):
    mask = data["geoId"]==pays[n]
    noms[n] = data[mask].reset_index(drop=True)
    #mask2 = data["deaths"] >= 100
    noms[n]["c_cumul"] = 0
    noms[n]["d_cumul"] = 0
    noms[n]["c_ratio"] = 0
    noms[n]["d_ratio"] = 0

    taille = len(noms[n])
    print("Nombre:", taille)

    for i in range(taille - 2, -1, -1):
        noms[n].loc[i, "c_cumul"] = noms[n].loc[i, "cases"] + noms[n].loc[i+1, "c_cumul"]
        noms[n].loc[i, "d_cumul"] = noms[n].loc[i, "deaths"] + noms[n].loc[i+1, "d_cumul"]

    for i in range(taille - 2, -1, -1):
        if noms[n].loc[i, "cases"] > 0:   # deja initialise à 0
            noms[n].loc[i, "c_ratio"] = noms[n].loc[i, "cases"] / noms[n].loc[i+1, "cases"]
        if noms[n].loc[i+1, "deaths"] > 0:   # deja initialise à 0
            noms[n].loc[i, "d_ratio"] = noms[n].loc[i, "deaths"] / noms[n].loc[i+1, "deaths"]

    # noms[n].head(3)
    # change the date format to a more suitable one for use

    noms[n]["Date"] = pd.to_datetime(noms[n]["dateRep"], dayfirst=True)
    #data_pays = data_pays.sort_values(["year", "month", "day"])
    noms[n] = noms[n].sort_values("Date")
    #noms[n].head(2)
```

```python
#PLOTS
fig, (ax1,ax2) = plt.subplots(1,2)
fig.figsize=(30,12)
for n in range(len(pays)):
    _= ax1.plot(noms[n]["Date"], noms[n]["c_cumul"], label=pays[n])
    _= ax2.plot(noms[n]["Date"], noms[n]["c_cumul"], label=pays[n])
ax2 = plt.yscale("log")
ax2 = plt.grid(True, which="both")
ax1 = plt.legend()
ax2 = plt.legend()
_= plt.title("Nombre de cas confirmés")
_= plt.show()
```
